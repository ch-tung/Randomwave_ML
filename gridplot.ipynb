```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams["font.family"] = "Arial"
from PIL import Image
```

```python
fig = plt.figure(figsize=(10, 10),dpi=192)

fz_list = np.arange(3)
fxy_list = np.arange(3)
fz_grid, fxy_grid = np.meshgrid(fz_list,fxy_list)

for f in zip(fz_grid.flatten(),fxy_grid.flatten()):
    ax = plt.subplot(3,3,(2-f[0])*3+f[1]+1)
    filename = './Figures/lamellar_{}-{}.png'.format(4*(f[0])+2,2*(f[1]+1))
    img = np.asarray(Image.open(filename))[25:575,25:575,:]
    ax.imshow(img)
    ax.axis('off')
    
plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```
