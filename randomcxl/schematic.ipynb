## Build Scene

This uses phase-winding vortex tracing for the two retained pairwise zero lines. Set `r.USE_VORTEX_TRACING = False` to return to the older `.contour()` tube-boundary visualization.

```python
plotter, meshes = r.build_scene()

print("S12 vortex points:", meshes["S12"].n_points, "lines:", meshes["S12"].n_lines)
print("S13 vortex points:", meshes["S13"].n_points, "lines:", meshes["S13"].n_lines)
print("candidate crosslink contacts:", len(meshes["crosslink_candidate_centers"]))
print("crosslink trace points:", meshes["crosslink_lines"].n_points, "lines:", meshes["crosslink_lines"].n_lines)
print("display crosslink site points:", meshes["display_crosslink_site_mesh"].n_points, "lines:", meshes["display_crosslink_site_mesh"].n_lines)
print("raw merged crosslink nodes:", len(meshes["raw_crosslink_centers"]))
print("adjusted crosslink nodes:", len(meshes["crosslink_centers"]))
print("displayed crosslink balls:", len(meshes["display_crosslink_centers"]))
```
